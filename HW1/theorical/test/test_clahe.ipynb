```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('testimg.png', 1)
```

```python
plt.imshow(image)
```

```python
lab_image = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
l, a, b = cv2.split(lab_image)
```

```python
plt.hist(l.flat, bins=256, range=(0,255), color='c')
```

```python
image.shape
```

```python
clahe = cv2.createCLAHE(clipLimit=200, tileGridSize =(9,9))
clahe_img = clahe.apply(l)
update_lab_image1 = cv2.merge((clahe_img, a, b))
CLAHE_image1 = cv2.cvtColor(update_lab_image1, cv2.COLOR_LAB2BGR)
```

```python
plt.imshow(CLAHE_image1)
```
